# Model Building

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,Y_train,Y_test = tts(features,target,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor as RFR
```

```python
reg_model = RFR()
reg_model.fit(X_train,Y_train)
```

```python
reg_model.score(X_train,Y_train) # Coefficient of determination R^2 of train data
```

```python
reg_model.score(X_test,Y_test) #Coefficient of determination R^2 of test data
```

# Hyperparametre Tuning

```python
RFR()
from sklearn.model_selection import RandomizedSearchCV as RSC
import numpy as np
```

```python
 #Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RFR()
```

```python
# Random search of parameters, using 5 fold cross validation, 
# search across 100 different combinations
rf_rndm = RSC(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 100, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_rndm.fit(X_train,Y_train)
```

```python
rf_rndm.best_params_
```

```python
rf_rndm.best_score_
```

```python
pred1 = rf_rndm.predict(X_test)
```

```python
sns.distplot(Y_test - pred1)
```

```python
plt.scatter(Y_test,pred1)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(Y_test, pred1))
print('MSE:', metrics.mean_squared_error(Y_test, pred1))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, pred1)))
```
